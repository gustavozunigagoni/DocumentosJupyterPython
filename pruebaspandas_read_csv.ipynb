### usecols

Use pandas usecols cuando desee cargar columnas específicas en el marco de datos. Cuando su conjunto de datos de entrada contiene una gran cantidad de columnas, y desea cargar un subconjunto de esas columnas en un marco de datos, entonces usecols será muy útil.

En cuanto al rendimiento, es mejor porque en lugar de cargar un marco de datos completo en la memoria y luego eliminar las columnas no necesarias, podemos seleccionar las columnas que necesitaremos, mientras cargamos el conjunto de datos en sí.

Como parámetro para usar los controles, puede pasar una lista de cadenas correspondientes a los nombres de columna o una lista de enteros correspondientes al índice de columna. Consulte el siguiente ejemplo donde paso una lista de cadenas como parámetro usecols.

[Contenido](#contenido)

```python
data = pd.read_csv(filepath_or_buffer  = fullpathfile, sep = ",",dtype = {"age":"float64","ticket":"object"},header =0, index_col = None,skip_blank_lines = True, na_filter = True)
data.info()
data = pd.read_csv(filepath_or_buffer  = fullpathfile, sep = ",",usecols =["name","age"])
data.head()
```

### infer_datetime_format

Si establece infer_datetime_format en True y habilita parse_dates para una columna, pandas read_csv intentará analizar el tipo de datos de esa columna en datetime rápidamente.

La ventaja de infer_datetime_format es que el análisis se realiza muy rápidamente.

Si tiene éxito en inferir el formato de fecha y hora de la cadena, la velocidad de análisis se incrementará en 10 veces.

### keep_date_col

A menudo, en ciencia de datos, trabajará con columnas de fecha para el análisis de datos.

Para comprender keep_date_col, consideremos que desea combinar las tres columnas, día, mes y año y derivar una nueva columna de fecha, llamada date_col

[Contenido](#contenido)

```python
nomarch = "fechas.txt"
filepath = "/Users/gustavozunigagoni/CursoDataSciencePython/MisDatasets"
fullpathfile = os.path.join(filepath,nomarch)
df = pd.read_csv(fullpathfile , parse_dates = {'date_col' : ["dia", "mes", "anno"]} )
 
df['date_col'] = pd.to_datetime(df['date_col'])
 
df.dtypes
```

Si establece keep_date_col en True, se conservarán las columnas de fecha originales, es decir, día, mes y año, junto con la nueva columna de fecha date_col en el marco de datos de pandas.

```python
nomarch = "fechas.txt"
filepath = "/Users/gustavozunigagoni/CursoDataSciencePython/MisDatasets"
fullpathfile = os.path.join(filepath,nomarch)
df = pd.read_csv(fullpathfile , parse_dates = {'date_col' : ["dia", "mes", "anno"]},keep_date_col = True)
df['date_col'] = pd.to_datetime(df['date_col'])
 
df.dtypes
```

### date_parser

Usar date_parser es la forma más flexible de analizar un archivo que contiene columnas de fecha

Suponga que tiene una columna llamada date_col en su conjunto de datos que tiene fecha en el formato AAAA DD MM HH: MM: SS, como se muestra a continuación

[Contenido](#contenido)

```python
nomarch = "fechas2.txt"
filepath = "/Users/gustavozunigagoni/CursoDataSciencePython/MisDatasets"
fullpathfile = os.path.join(filepath,nomarch)
pd.read_csv(fullpathfile)
```

La forma más fácil es escribir una función lambda que pueda leer los datos en este formato y pasar esta función lambda al argumento date_parser.

```python
mydateparser = lambda x: pd.datetime.strptime(x, "%Y %m %d %H:%M:%S")
nomarch = "fechas2.txt"
filepath = "/Users/gustavozunigagoni/CursoDataSciencePython/MisDatasets"
fullpathfile = os.path.join(filepath,nomarch)
pd.read_csv(fullpathfile, parse_dates = ['fecha'],date_parser = mydateparser)
```

### chunksize

Al especificar un tamaño de fragmento, puede recuperar los datos en "fragmentos" del mismo tamaño.

Esto es especialmente útil al leer un gran conjunto de datos como parte de su proyecto de ciencia de datos.

En lugar de poner todo el conjunto de datos en la memoria, esta es una forma "perezosa" de leer porciones de datos del mismo tamaño. Establecer chunksize devolverá un objeto TextFileReader.

Vea un ejemplo a continuación, he especificado un tamaño de fragmento de 2 aquí. Por lo tanto, se devolverán dos filas para cada ejecución de bucle.

[Contenido](#contenido)

```python
temp= u"""fecha_inicio
04/10/96
15/07/97
10/12/97
06/10/99
//1994
/02/67
"""
data = pd.read_csv(io.StringIO(temp),parse_dates = ['fecha_inicio'], dayfirst = True, chunksize = 2)
```

```python
for chunk in data:
    print(chunk)
```

### compression

pandas read_csv tiene la capacidad de leer archivos comprimidos.

De forma predeterminada, el parámetro de compresión se establece en "inferir", lo que significa que intentará descomprimir los archivos del tipo (gzip, zip, bz2, xz) y leerlos en el marco de datos.

Si desea hacer un análisis en un archivo enorme, siempre es mejor usar un archivo comprimido.

No descomprima el archivo e intente leerlo en el marco de datos.

Es decir, si tiene un archivo gz, configure la compresión = "gzip" y trabaje con el archivo gz directamente.

### thousands

Si una columna dentro de su conjunto de datos contiene una coma para indicar el lugar de los miles, e intenta convertir este conjunto de datos en un marco de datos usando pandas read_csv, ¡entonces esta columna se consideraría como una cadena!

[Contenido](#contenido)

```python
temp= u"""monto
1,234,556.5
13,514.50
34,607
55,789
1,555.10
3,459.90
"""
data = pd.read_csv(io.StringIO(temp),sep = ';')
data.dtypes
```

```python
temp= u"""monto
1,234,556.5
13,514.50
34,607
55,789
1,555.10
3,459.90
"""
data = pd.read_csv(io.StringIO(temp),sep = ';',thousands =',')
data.dtypes
```
